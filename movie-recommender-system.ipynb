```python
#Countvectorizer used to transform a given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text,
#Convert a collection of text documetns to a matrix of token counts
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#Learn the vocabulary dictionary and return document-term matrix
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
#Cosine distance to calculate the similarity of tags
#sklearn provides a class for calculating pairwise cosine_similarity
```

```python
#Similarity refers to distance with dimensions representing features of data object
#in dataset, if the distance is less there will be a high degree of similarity
#Cosine similarity is a metric hellpful in determingn how similar the data objects are
#irrespective of their size. Formula:cos(x,y)-x.y/||x||*||y||
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#Enumerate: function used to return both indexes and values of the list
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avengers: Age of Ultron")
```
